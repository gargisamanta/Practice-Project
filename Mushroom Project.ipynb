```python
# Counts of all the features

All_cols= df_mushroom.columns.to_list()
print ('counts:')
for col in All_cols:
    print(col,'\n',df_mushroom[col].value_counts())
```

# Data Visualization

```python
# Compare each feature with respect to class

for col in All_cols:
    plt.figure(figsize=(20,5))
    plt.subplot(2,3,1)
    sns.countplot(x=col,hue='class',data=df_mushroom)
    plt.xlabel(col,fontsize=12)
    plt.legend(loc='upper right')
    plt.show()
```

```python
# On the basis of visualization we can delete some of features : class,gill-attachment,stalk-shape, stalk=surface-above-ring,
# stalk-color-above-ring, viel-color, spore-print-color,ring-number
# So we can create a list of deleted columns and from from the main dataset
# Apply One Hot Encoding on the features

del_cols=['gill-attachment','stalk-shape', 'stalk-surface-above-ring','stalk-color-above-ring','veil-color',
         'spore-print-color','ring-number']

for col in del_cols:
    df_mushroom=df_mushroom.drop(col,axis=1)
df_mushroom=pd.get_dummies(df_mushroom,columns=df_mushroom.columns,drop_first=True)
df_mushroom.head()


```

```python
df_mushroom.shape
```

# Model buiding

```python
y=df_mushroom['class_p']
x=df_mushroom.drop('class_p',axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=800)
```

```python
print('X-train shape:',x_train.shape)
print('Y-train shape:',y_train.shape)
print('X-test shape:',x_test.shape)
print('Y-test shape:',y_test.shape)
```

```python
# As it is clearly a classification problem we will be using
# Logistic Regression
# Decision Tree Classifier
# Random Forest Classifier
# XGBoost Classifier
```

```python
# Logistic Regression

from sklearn.linear_model import LogisticRegression
```

```python
lgr=LogisticRegression()
```

```python
lgr.fit(x_train,y_train)
```

```python
lgr.score(x_test,y_test)
```

```python
y_pred=lgr.predict(x_test)
```

```python
confusion_matrix(y_test,y_pred)
```
