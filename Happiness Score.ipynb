```python
# As we will be considering region so we are dropping the country columns

df.drop(['Country'],inplace=True,axis=1)
```

```python
df.shape
```

```python
# Count the values of region
df['Region'].value_counts()
```

```python
# Change the Ctegorical variable,region into numerical data using ONE HOT ENCODING

region=pd.get_dummies(df['Region'],drop_first=True)
region
```

```python
# CONCAT THE NEW COLUMNS TO THE DATASET

df=pd.concat([df,region],axis=1)
df
```

```python
df.shape
```

```python
# Drop the original column from the dataset

df.drop(['Region'],inplace=True,axis=1)
```

```python
df.shape
```

```python

df.head()
```

# MODEL BUILDING

```python
y=df['Happiness Score']
x=df.drop('Happiness Score',axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=300)
```

```python
print('X-train shape:',x_train.shape)
print('Y-train shape:',y_train.shape)
print('X-test shape:',x_test.shape)
print('Y-test shape:',y_test.shape)
```

# Train the Model using Multiple Regressor

```python
# Import the model

from sklearn.linear_model import LinearRegression
```

```python
# create an object
lr=LinearRegression()
```

```python
# Fit the model
lr.fit(x_train,y_train)
```

# Find the score

```python
lr.score(x_test,y_test)
```

```python
# Predict the value with any test value

y_pred=lr.predict(x_test)
y_pred
```

```python
# Finding the accuracy using MSE
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,y_pred)
```

# Train the Model using Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dtr=DecisionTreeRegressor()
```

```python
dtr.fit(x_train,y_train)
```

```python
# Find the score

dtr.score(x_test,y_test)
```

```python
y_pred1=dtr.predict(x_test)
y_pred1
```

```python
mean_squared_error(y_test,y_pred1)
```

# Train the Model using Random forest Regressor 

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr=RandomForestRegressor()
```

```python
rfr.fit(x_train,y_train)
```

```python
rfr.score(x_test,y_test)
```

```python
y_pred2=rfr.predict(x_test)
y_pred2
```

```python
mean_squared_error(y_test,y_pred2)
```
